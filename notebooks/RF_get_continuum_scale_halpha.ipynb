```python
# plot halpha FWHM vs R

x = v.halpha['R_FWHM']
y = v.halpha['H_FWHM']

flag = v.main['HAflag']

#plt.scatter(x[flag],y[flag],c=v.halpha['TEL'][flag])

tels = set(v.halpha['TEL'][flag])
for t in tels:
    flag1 = v.halpha['TEL'] == t
    plt.scatter(x[flag&flag1],y[flag&flag1],s=50,label=t,alpha=.5)
    diff = (x[flag&flag1] - y[flag&flag1])/([x[flag&flag1]])
    print(f"{t} rms = {np.std(diff):.2f}")
    
plt.legend()

xline = np.linspace(1,3,100)
plt.plot(xline,1.1*xline,'k--')
plt.plot(xline,0.9*xline,'k--')
plt.xlabel("R-band FWHM")
plt.xlabel("H-alpha FWHM")
```

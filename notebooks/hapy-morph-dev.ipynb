halpha morphologies are currently measured using a SNR threshold, but this means the metrics will return different values for the same galaxy if it falls in an image with different sky noise.

the same is true for the segmentation region defined for the r-band image - we are requiring snr>1.5 for 10 pixels, so the same problem applies.


Potential solutions

- set sb threshold of r-band image based on petrosian radius
- the set sb threshold of halpha image based on an abosolute flux limit.
